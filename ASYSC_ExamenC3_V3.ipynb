```python
# importando math para la raiz, cossignal para la multiplicación por coseno y thinkplot para decorate
import math
from thinkdsp import CosSignal
import thinkplot
```

```python
# variables
# casos Solo cambiando valores de R, L y C
vm = 5 
w = 2000 
B = 0
R = 10   #Casos: 1;10     2;1000   3;100    4;100     5;50
L = 0.01 #Casos: 1;0.01   2;0.01   3;0.05   4;0.01   5;0.01
C = 1e-6 #Casos: 1;1e-6   2;1e-6   3;1e-6   4;10e-6   5;1e-6
```

```python
# hacemos operaciones una por una para mas facil
# parte de theta
A = w * L - 1/(w * C) #para no reescribir tanto ;D

# modulo de z
Zjw = math.sqrt(R**2 + A**2)
#theta de t
arct = math.atan2(A,R)
# i(t)
i1 = vm / Zjw # amplitud
#--------------------------------------
phi = B - arct # fase final

f = w / (2*math.pi) #thinkdsp usa Hz, la frecuencia
print("f= ",f) #la pide la tabla del examen
signal = CosSignal(amp=i1, freq=f, offset=phi) #"señal"

wave = signal.make_wave(duration=0.1, framerate=1000) #la onda pa graficar
#-------------------------------------------
```

```python
# se grafica
wave.plot()
thinkplot.show()
```

```python
# ahora el especto
spectrum = wave.make_spectrum()
spectrum.plot()
thinkplot.show()
```

```python
# la fase
thinkplot.plot(spectrum.fs, spectrum.angles)
thinkplot.show()
```
